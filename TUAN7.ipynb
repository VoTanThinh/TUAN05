```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression
data = pd.read_csv('50_Startups.csv')
data.head(0)
#X = data.drop(data.columns[[0,3]], axis = 1)
#-------------------------------------------#
sns.pairplot(data)
#-------------------------------------------#
corr = data.corr()
sns.heatmap(corr, annot = True)
#-------------------------------------------#
X = data.drop(['Profit'], axis = True)
y = data['Profit']
X.head()
#-------------------------------------------#
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
#-------------------------------------------#
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.5, random_state = 0)

X_train.shape
X_test.shape
y_train.shape
y_test.shape

regressor = LinearRegression()
regressor.fit(X_train,y_train) #gọi hàm cho nó học
y_pred=regressor.predict(X_test).round(1) #tuyên đón làm tròn với 1 số
calculation=pd.DataFrame(np.c_[y_test,y_pred],columns=['Original Profit','Predicted Profit'])
print(calculation)
```
